****Import modules****

```python
import os
import numpy as np
```

****Path of the directory of the corpus****

```python
path = 'data/text_imdb'
```

```python
filenames = os.listdir(path)
filenames
```

****Load corpus****

```python
corpus = []
chars_to_remove = ['\'', '<br', '/>', ',', '.', '!', '--', '(', ')', '-']
for filename in filenames:
    f = open(path + '/' + filename)
    lines = f.readlines()
    f.close()
    lines = lines[0]
    for char_to_remove in chars_to_remove:
        if char_to_remove == '-':
            lines = lines.replace(char_to_remove, ' ')
        else:
            lines = lines.replace(char_to_remove, '')
    corpus.append(lines)
```

```python
corpus
```

****Function for building the word index dictionary (i.e. vocabulary)****

```python
def create_word_index(corpus):
    word_index = {}
    for text in corpus:
        words = text.split(' ')
        for word in words:
            if word not in word_index:
                word_index[word] = len(word_index)
    return word_index
```

****Function for calculating the maximum number of words per text in the entire corpus****

```python
def max_num_of_words_per_text(corpus):
    max_num_of_words = -1
    for text in corpus:
        words = text.split(' ')
        num_of_words = len(words)
        if num_of_words > max_num_of_words:
            max_num_of_words = num_of_words
    return max_num_of_words
```

****Function for tranforming the initial corpus (i.e. list of strings) into a 3D numerical array of shape (max_num_of_words, vocab_size, num_of_texts_in_corpus) using the one-hot-encoding (OHE) technique****

```python
def word_vectorize_ohe(corpus):
    num_of_texts_in_corpus = len(corpus)
    word_index = create_word_index(corpus)
    vocab_size = len(word_index)
    max_num_of_words = max_num_of_words_per_text(corpus)
    word_vectors = np.zeros((max_num_of_words, vocab_size, num_of_texts_in_corpus))
    for k, text in enumerate(corpus):
        words = text.split(' ')
        for i, word in enumerate(words):
            j = word_index.get(word)
            word_vectors[i, j, k] = 1
    return word_vectors
```

```python
word_vectors = word_vectorize_ohe(corpus)
word_index = create_word_index(corpus)
```

```python
text = corpus[0]
text = text.split(' ')
word_vector = word_vectors[:,:,0]
for i, word in enumerate(text):
    print(word, word_index.get(word),word_vector[i,:])
```
